```python
data = ModelData.OpenData()
```

```python
import matplotlib.pyplot as plt

# Determine all time indices you want to plot
timeIndices = range(0, data.sizes["time"], 10)

fig, ax = plt.subplots()

for t in timeIndices:
    rho = data.isel(time=t)["rho"]
    winterp = data.isel(time=t)["winterp"]
    rhow = rho * winterp

    # Matplotlib automatically assigns a different color for each plot call
    ax.plot(rhow.mean(dim=("xh", "yh")), ModelData.zh, label=f"time={t}")

ax.set_xlabel("rho*w (kg/m^3 * m/s)")
ax.set_ylabel("z (km)")
plt.title("Average Vertical Profiles")
ax.legend()
```
