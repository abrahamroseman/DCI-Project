```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
path='../../../Model/cm1r20.3/run/MODEL_OUTPUT/Simulation_Two/cm1out_000100.nc'
data=xr.open_dataset(path,decode_timedelta=False)

varName='thflux'
a = data[varName].isel(time=0)
a -=a.mean()
cmap='RdBu_r'
plt.contourf(a,cmap=cmap);plt.colorbar()
```
